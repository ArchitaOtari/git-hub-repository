```python
from sklearn.datasets import load_iris
import pandas as pd

i = load_iris()
df = pd.DataFrame(i.data, columns=i.feature_names)
df['target'] = i.target
```

```python
from matplotlib import pyplot as plt
df['sepal length (cm)'].plot(kind='hist', bins=0, title='sepal length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot([
df['sepal length (cm)'][df['target']==0],
df['sepal length (cm)'][df['target']==1],
df['sepal length (cm)'][df['target']==2]
],
labels=['Target 0','Target 1','Target 2']
)
plt.xlabel('Target Class')
plt.ylabel('Sepal Length (cm)')
_=plt.title('Distribution of Sepal Length by Target Class')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title(' Correlation between Sepal Length and Sepal Width')
_=plt.colorbar(label='Target Class')
```

```python
import seaborn as sns

species_counts = df['target'].value_counts()

plt.figure(figsize=(8,5))
sns.set_palette('pastel')
explode= [0.2, 0, 0]
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Peoportion  of Iris Species')
plt.legend()
plt.show
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='target', palette='plasma')
plt.title('Iris Dataset scatter plot matrix')
plt.show()
```
