```python
from sklearn.datasets import load_iris
import pandas as pd

i = load_iris()
df = pd.DataFrame(i.data, columns=i.feature_names)
df['target'] = i.target



```

```python
from matplotlib import pyplot as plt
df['sepal length (cm)'].plot(kind='hist', bins=30, title='sepal length (cm)')
plt.show()
```
